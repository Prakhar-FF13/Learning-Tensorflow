# Convolutional Neural Networks :

Dataset used - MNIST

```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
import matplotlib.pyplot as plt
```

```python
batch_size = 128
num_classes = 10
epochs = 10
```

```python
img_rows, img_cols = 28, 28
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
if K.image_data_format() == 'channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

### A) Model 1:

1. First convolutional layer with 32 filters, kernel size of 3X3, and no zero padding, this will reduce the size.
2. Second Convolutional layer with 64 filters and kernel size of 3X3 and no zero padding.
3. Another Convolutional layer with 128 filters and kernel size of 3X3
4. Max Pooling Layer with matrix size of 2X2 and padding to keep the size same.
5. A dense layer of 64 neurons
6. A dense layer of 10 neurons for output

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape= input_shape ))
model.add(Conv2D(64, kernel_size=(3,3), activation='relu') )
model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), padding="same"))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'r',linewidth=3.0)
plt.plot(history.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

1. Looking at the loss curve we can see that the differnce between train and test loss has started to increase from epoch 1 this could mean overfitting.
2. The accuracy on both the train and test sets are close at around 99.99% for train and 98% for test

### B) Model 2:

1. First convolutional layer with 32 filters, kernel size of 3X3, and no zero padding, this will reduce the size.
2. Second Convolutional layer with 64 filters and kernel size of 3X3 and no zero padding.
3. Adding a dropout of 0.2
4. Max Pooling Layer with matrix size of 2X2.
5. A dense layer of 64 neurons
6. A dense layer of 10 neurons for output

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape= input_shape ))
model.add(Conv2D(64, kernel_size=(3,3), activation='relu') )
model.add(Dropout(0.2))
model.add(MaxPooling2D(pool_size=(2,2), padding="same"))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'r',linewidth=3.0)
plt.plot(history.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

1. Here also we see the same story, the model has started to overfit as the seperation between the loss curves of train and test both keep increasing but the accuracy here keeps on increasing for both the sets which is contradictory as generally more overfitting causes less accuracy on test data.

### C) Model 3:

1. First convolutional layer with 64 filters, kernel size of 5X5.
2. Adding a dropout of 0.3
3. Second Convolutional layer with 128 filters and kernel size of 2X2 and no zero padding.
4. Max Pooling Layer with matrix size of 5X5.
5. Adding a dropout of 0.2.
6. A dense layer of 64 neurons
7. A dense layer of 10 neurons for output

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(5,5), activation='relu', padding = 'same', input_shape= input_shape ))
model.add(Dropout(0.3))
model.add(Conv2D(128, kernel_size=(2,2), activation='relu') )
model.add(MaxPooling2D(pool_size=(5,5)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'r',linewidth=3.0)
plt.plot(history.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

1. This model is different than the previous two, here there is not significant gap between train and test loss curves, except the last few epochs where it may seem that model just began overfitting.
2. Accuracy also seems to improve as the epochs increase

### D) Model 4:

1. First convolutional layer with 8 filters, kernel size of 9X9. Padding same.
2. Conv layer 16 filters kernel size of 7X7 Padding Same.
3. Dropout of 0.4
4. MaxPool 5X5 Padding same.
5. Conv layer 32 filters kernel size 5X5
6. Conv Layer 64 Filters kernel size 3X3 padding same.
7. Conv Layer 128 Filters kernel size 2X2 Padding same.
7. Dropout of 0.3
8. MaxPool 3X3 Padding same.
9. A dense layer of 128 neurons
10. A dense layer of 10 neurons for output

```python
model = Sequential()
model.add(Conv2D(8, kernel_size=(9,9), activation='relu', padding = 'same', input_shape= input_shape ))
model.add(Conv2D(16, kernel_size=(7,7), activation='relu', padding = 'same') )
model.add(Dropout(0.4))
model.add(MaxPooling2D(pool_size=(5,5), padding = 'same', strides=(1,1)))
model.add(Conv2D(32, kernel_size=(5,5)))
model.add(Conv2D(32, kernel_size=(3,3), padding = 'same'))
model.add(Conv2D(32, kernel_size=(2,2)))
model.add(Dropout(0.3))
model.add(MaxPooling2D(pool_size=(2,2) , padding = 'same'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'r',linewidth=3.0)
plt.plot(history.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

1. For this model too, the loss curves seperate as the epochs go on.

### E) Model 5:

1. First convolutional layer with 20 filters, kernel size of 2X2. Padding same.
2. Dropout of 0.6
2. Conv layer 40 filters kernel size of 3X3 Padding Same.
3. Dropout of 0.5
4. Conv Layer with 60 Filters kernel sie of 4X4 padding same.
5. Dropout of 0.4
5. Conv layer 80 filters kernel size 5X5 padding same.
6. Dropout of 0.3
7. Conv Layer 100 Filters kernel size 4X4 padding same.
8. Dropout of 0.2
9. Conv Layer 120 Filters kernel size 3X3
10. Dropout of 0.1
11. Conv Lyer with 140 Filters kernel size of 2X2
12. Max Pooling of 4X4.
9. A dense layer of 128 neurons
10. A dense layer of 10 neurons for output

```python
model = Sequential()
model.add(Conv2D(20, kernel_size=(2,2), activation='relu', padding = 'same', input_shape= input_shape ))
model.add(Dropout(0.6))
model.add(Conv2D(40, kernel_size=(3,3), activation='relu', padding = 'same') )
model.add(Dropout(0.5))
model.add(Conv2D(60, kernel_size=(4,4), activation ='relu', padding ='same') )
model.add(Dropout(0.4))
model.add(Conv2D(80, kernel_size=(5,5), padding = 'same', activation = 'relu'))
model.add(Dropout(0.3))
model.add(Conv2D(100, kernel_size=(4,4), padding = 'same', activation = 'relu'))
model.add(Dropout(0.2))
model.add(Conv2D(120, kernel_size=(3,3), activation = 'relu'))
model.add(Dropout(0.1))
model.add(Conv2D(140, kernel_size=(2,2), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(4,4) , padding = 'same', strides=(1,1)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```python
model.evaluate(x_test, y_test)
```

```python
# Loss Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['loss'],'r',linewidth=3.0)
plt.plot(history.history['val_loss'],'b',linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.title('Loss Curves',fontsize=16)
 
# Accuracy Curves
plt.figure(figsize=[8,6])
plt.plot(history.history['acc'],'r',linewidth=3.0)
plt.plot(history.history['val_acc'],'b',linewidth=3.0)
plt.legend(['Training Accuracy', 'Validation Accuracy'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.title('Accuracy Curves',fontsize=16)
```

1. This model is clearly not overfitting as the accuracy seems to be increasing with every epoch and also the loss curves do not have much seperation.
